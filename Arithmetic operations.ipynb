Arithmetic Operations

```python
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var
```

```python
var["C"] = var["A"]+var["B"]
var
```

```python
var["D"] = var["A"]-var["B"]
var
```

```python
var["E"] = var["A"]*var["B"]
var
```

```python
var["F"] = var["A"]/var["B"]
var
```
